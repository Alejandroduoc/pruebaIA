```python
# Importar bibliotecas necesarias
from langchain_openai import ChatOpenAI

from langchain.schema import HumanMessage, AIMessage, SystemMessage
import os
import time

print("Bibliotecas importadas correctamente para streaming")

try:
    llm = ChatOpenAI(
        base_url="https://models.github.ai/inference",  #descubri que aqui  cambia de url base  
        api_key=os.getenv("GITHUB_TOKEN"),               
        model = "openai/gpt-4o-mini",               
        temperature=0.7,
        streaming=True  # Sin streaming
       
    )

    
    print("âœ“ Modelo configurado con streaming habilitado")
    print(f"Modelo: {llm.model_name}")
    print(f"Streaming: {llm.streaming}")
    
except Exception as e:
    print(f"âœ— Error en configuraciÃ³n: {e}")
    print("Verifica las variables de entorno")

```

```python



conversation_history = []
current_personality = "amigable"

personalities = {
    "amigable": "Eres un asistente amable y simpÃ¡tico.",
    "formal": "Eres un asistente profesional y formal.",
    "divertido": "Eres un asistente divertido y jovial."
}

def print_help():
    print("Comandos especiales:\n/help - Muestra esta ayuda\n/clear - Limpia el historial\n/personality [nombre] - Cambiar personalidad")

print("Chatbot Mejorado. Escribe '/help' para ayuda.")

while True:
    user_input = input("Usuario: ").strip()
    if user_input.lower() == "salir":
        print("Hasta luego!")
        break

    if user_input.startswith("/"):
        if user_input == "/help":
            print_help()
            continue
        elif user_input == "/clear":
            conversation_history.clear()
            print("Historial limpiado.")
            continue
        elif user_input.startswith("/personality"):
            _, _, p = user_input.partition(" ")
            if p in personalities:
                current_personality = p
                print(f"Personalidad cambiada a {p}")
            else:
                print(f"Personalidad desconocida. Opciones: {list(personalities.keys())}")
            continue
        else:
            print("Comando no reconocido. Escribe /help para ayuda.")
            continue

    # Construir la conversaciÃ³n con el mensaje del sistema de la personalidad actual
    messages = [SystemMessage(content=personalities[current_personality])] + conversation_history + [HumanMessage(content=user_input)]

    full_response = ""
    print("Bot:", end=" ", flush=True)
    try:
        for chunk in llm.stream(messages):
            print(chunk.content, end="", flush=True)
            full_response += chunk.content
        print()
    except Exception as e:
        print(f"\nError: {e}")

    # Guardar interacciÃ³n en el historial
    conversation_history.append(HumanMessage(content=user_input))
    conversation_history.append(AIMessage(content=full_response))
```

```python
def chatbot_streaming():
    # Muestra tÃ­tulo inicial en la consola
    print("=== CHATBOT CON STREAMING MEJORADO ===")
   

    # Instrucciones bÃ¡sicas para salir de la conversaciÃ³n
    print("Escribe 'salir' para terminar la conversaciÃ³n\n")
    # InstrucciÃ³n para ver ayuda con comandos especiales
    print("Escribe '/help' para pedir ayuda\n")

   


    # DefiniciÃ³n de personalidades que el chatbot puede adoptar.
    personalities = {
        "amigable": """Eres un asistente Ãºtil y amigable especializado en tecnologÃ­a. 
                      Respondes de manera clara y concisa, y siempre intentas ser educativo.""",
        "formal": "Eres un asistente profesional y formal, que responde con respeto y precisiÃ³n.",
        "divertido": "Eres un asistente divertido y jovial, que responde con humor y energÃ­a positiva."
    }
    print("Personalidades disponibles:", ", ".join(personalities.keys()))
    # Variable que almacena la personalidad activa, por defecto "amigable"
    current_personality = "amigable"
    # Mensaje del sistema que define el comportamiento segÃºn personalidad
    system_message = personalities[current_personality]

    # Lista que mantiene todo el historial de conversaciÃ³n para que el modelo tenga contexto
    conversation_history = []

    # FunciÃ³n interna para mostrar comandos y ayuda al usuario
    def print_help():
        print("\nComandos especiales:")
        print(" /help - Muestra esta ayuda")  # Muestra este mensaje
        print(" /clear  - Limpia el historial de conversaciÃ³n")  # Limpia mensajes guardados
        print(" /personalidad X   - Cambia la personalidad del asistente a X (amigable, formal, divertido)")  # Cambia personalidad
        print(" salir   - Termina la conversaciÃ³n\n")  # Termina el chat

    # Bucle principal infinito para recibir input y responder
    while True:
        # Lee input del usuario, elimina espacios extras al inicio y final
        user_input = input("\n TÃº: ").strip()

        # Si usuario quiere salir, muestra mensaje de despedida y rompe el bucle
        if user_input.lower() in ['salir', 'exit', 'quit']:
            print("\n Â¡Hasta luego!")
            break

        # Si la entrada estÃ¡ vacÃ­a (solo espacios), se ignora y vuelve a leer input
        if not user_input:
            continue

        # Si input empieza con "/" se interpreta como comando especial
        if user_input.startswith("/"):
            # Comando /help muestra ayuda
            if user_input == "/help":
                print_help()
                continue
            # Comando /clear limpia el historial guardado
            elif user_input == "/clear":
                conversation_history.clear()
                print("ðŸ§¹ Historial limpiado.")
                continue
            # Comando para cambiar personalidad, recibe argumento tras espacio
            elif user_input.startswith("/personalidad"):
                _, _, p = user_input.partition(" ")  # Separa comando y argumento
                # Verifica si la personalidad existe en diccionario
                if p in personalities:
                    current_personality = p  # Cambia personalidad activa
                    system_message = personalities[current_personality]  # Actualiza mensaje sistema
                    conversation_history.clear()  # Limpia historial porque cambia contexto
                    print(f" Personalidad cambiada a '{p}'. Historial reiniciado.")
                else:
                    print(f" Personalidad desconocida. Opciones: {list(personalities.keys())}")
                continue
            else:
                # Si comando no reconocido, indica error y muestra ayuda
                print(" Comando no reconocido. Escribe /help para ayuda.")
                continue

        # Mensaje que indica que el asistente va a responder
        print("\n Asistente: ", end="", flush=True)

        try:
            # Construye el listado de mensajes para enviar al modelo:
            # primero mensaje sistema con la personalidad, luego toda la historia,
            # y finalmente el nuevo mensaje del usuario
            messages = [SystemMessage(content=system_message)] + conversation_history + [HumanMessage(content=user_input)]

            full_response = ""  # Variable para guardar la respuesta completa del asistente

            # Streaming para recibir la respuesta token a token y mostrarla en tiempo real
            for chunk in llm.stream(messages):
                print(chunk.content, end="", flush=True)  # Muestra contenido parcial en consola
                full_response += chunk.content  # Acumula parcial en la respuesta completa
                time.sleep(0.02)  # PequeÃ±a pausa para dar efecto streaming mÃ¡s natural

            # Salto de lÃ­nea para finalizar la impresiÃ³n de la respuesta
            print()

            # Guarda el mensaje del usuario y la respuesta del asistente en el historial
            conversation_history.append(HumanMessage(content=user_input))
            conversation_history.append(AIMessage(content=full_response))

        # Captura interrupciÃ³n con Ctrl+C para terminar limpia y amigablemente
        except KeyboardInterrupt:
            print("\n Interrumpido por el usuario")
            break
        # Captura cualquier error inesperado y lo muestra
        except Exception as e:
            print(f"\n Error: {e}")

    # Mensaje final cuando termina la conversaciÃ³n
    print("\nÂ¡Gracias por usar el chatbot!")


# Para probar, descomenta la siguiente lÃ­nea:
chatbot_streaming()
```
